```python
df = df.assign(net_ion_pair = lambda x: x['Ion generation']-x['Ion recombination'] )
```

# Calculate ion pairs availability coefficient

```python
max_ion_pair = df['net_ion_pair'].max()
net_ion_pair = df['net_ion_pair'].to_list()
time = df.iloc[:,0].to_list()
index = list(range(len(time)-1))

#total simulation time
T_sim = df.iloc[-1,0]
```

```python
ion_lifetime = []

while max_ion_pair >= 0:
    ion_time.append([])
    max_ion_pair -= 1
    
for i in index: 
    ion_time[net_ion_pair[i]].append(time[i+1]-time[i])
    
sum_ion_lifetime = []

for i in ion_time:
    sum_ion_lifetime.append(sum(i))
```

```python
sum(ion_lifetime[1:])
sum(ion_lifetime[1:])/T_sim
```
